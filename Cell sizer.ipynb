```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Import an image.

```python
src = cv2.imread('img/8 bit mcbl 125/1/8-Bit Copy of Acquired-3.tif', cv2.IMREAD_COLOR)
plt.imshow(src)
```

http://docs.opencv.org/trunk/modules/imgproc/doc/filtering.html

C++: Mat getGaborKernel(Size ksize, double sigma, double theta, double lambd, double gamma, double psi=CV_PI*0.5, int ktype=CV_64F )

Python: cv2.getGaborKernel(ksize, sigma, theta, lambd, gamma[, psi[, ktype]]) → retval
*  Parameters:
        ksize – Size of the filter returned.
        sigma – Standard deviation of the gaussian envelope.
        theta – Orientation of the normal to the parallel stripes of a Gabor function.
        lambd – Wavelength of the sinusoidal factor.
        gamma – Spatial aspect ratio.
        psi – Phase offset.
        ktype – Type of filter coefficients. It can be CV_32F or CV_64F .

```python
class Gabor(object):
    def __init__(self, sigma, lambd, gamma, psi, shape=src.shape[-2::-1]):
        self._shape = shape
        self._sigma = sigma
        self._lambd = lambd
        self._gamma = gamma
        self._psi = psi
        self.update()
        
    def __str__(self):
        return """\
        Sigma: {}
        Lambda: {}
        Gamma: {}
        Psi: {}
        """.format(self._sigma, self._lambd, self._gamma, self._psi)
        
    def sigma(self, s):
        self._sigma = s
        self.update()
        
    def lambd(self, l):
        self._lambd = l
        self.update()
        
    def gamma(self, g):
        self._gamma = 2 * np.pi * g / 360.0
        self.update()
        
    def psi(self, p):
        self._psi = 2 * np.pi * p / 360.0
        self.update()
    
    def update(self):
        kern = cv2.getGaborKernel(self._shape, self._sigma, np.pi/2, self._lambd, self._gamma, self._psi, ktype=cv2.CV_32F)
        self.img = cv2.filter2D(src, -1, kern)
```

```python
cv2.namedWindow('image')

gabor = Gabor(0,0,0,0)

cv2.createTrackbar('sigma', 'image', 0, 20, gabor.sigma)
cv2.createTrackbar('lambda', 'image', 0, 200, gabor.lambd)
cv2.createTrackbar('gamma', 'image', 0, 360, gabor.gamma)
cv2.createTrackbar('psi', 'image', 0, 360, gabor.psi)

k = 0xFF
while k != 27:
    cv2.imshow('image', gabor.img)
    k = cv2.waitKey(1) & 0xFF
        
else:
    cv2.destroyAllWindows()
    print(gabor)
```

```python
kern = cv2.getGaborKernel(src.shape[-2::-1], 3, theta=np.pi/2, lambd=200, gamma=3.68, psi=1.52, ktype=cv2.CV_32F)
#plt.imshow(kern)
filtered = cv2.filter2D(src, -1, kern)
plt.imshow(filtered)
```

```python
ret, thresh = cv2.threshold(filtered, 90, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
```

```python
mask = thresh.copy()
cv2.floodFill(mask, None, (1, 1), (255,255,255))
fill = cv2.bitwise_xor(thresh, cv2.bitwise_not(mask))
plt.imshow(fill)
```
